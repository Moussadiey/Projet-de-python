## <strong>Premiers pas</strong>
</div>

Le document que vous consultez n'est pas une page Web statique, mais un environnement interactif appelé <strong>notebook Colab</strong>, qui vous permet d'écrire et d'exécuter du code.

Voici par exemple une <strong>cellule de code</strong> avec un bref script en Python qui calcule une valeur, l'enregistre dans une variable et imprime le résultat :

```
seconds_in_a_day = 24 * 60 * 60
seconds_in_a_day
```

Pour exécuter le code dans la cellule ci-dessus, sélectionnez-le en cliquant dessus, puis cliquez sur le bouton de lecture à gauche du code ou utilisez le raccourci clavier Commande/Ctrl+Entrée. Pour modifier le code, cliquez simplement sur la cellule.

Les variables que vous définissez dans une cellule peuvent être utilisées par la suite dans d'autres cellules :

```
seconds_in_a_week = 7 * seconds_in_a_day
seconds_in_a_week
```

Les notebooks Colab vous permettent d'utiliser, dans un même document, du <strong>code exécutable</strong>, du <strong>texte enrichi</strong>, des <strong>images</strong>, du code <strong>HTML</strong>, du code <strong>LaTeX</strong> et bien plus. Traitement de données LiDAR pour la détection des sommets d'abres par la couche MNH

```
import rasterio
from skimage.feature import peak_local_max
from skimage.segmentation import watershed
from scipy import ndimage
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path

# === 1) Charger le CHM raster ===
chm_path = "/content/chemin/vers/MNH.tif"
with rasterio.open(chm_path) as src:
    chm = src.read(1)
    profile = src.profile  # Pour réutiliser les métadonnées

# === 2) Nettoyer les valeurs négatives (si besoin) ===
chm[chm < 0] = 0

# === 3) Calculer la transformée de distance ===
distance = ndimage.distance_transform_edt(chm > 0)

# === 4) Détecter les maxima locaux (nouvelle syntaxe) ===
coordinates = peak_local_max(
    distance,
    footprint=np.ones((1, 1)),
    labels=(chm > 0)
)

local_maxi = np.zeros_like(distance, dtype=bool)
local_maxi[tuple(coordinates.T)] = True

# === 5) Créer les marqueurs ===
markers, _ = ndimage.label(local_maxi)

# === 6) Appliquer Watershed ===
labels = watershed(-distance, markers, mask=(chm > 0))

# === 7) Afficher ===
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

axes[0].imshow(chm, cmap='terrain')
axes[0].set_title("CHM")

axes[1].imshow(markers, cmap='nipy_spectral')
axes[1].set_title("Marqueurs (sommets détectés)")

axes[2].imshow(labels, cmap='nipy_spectral')
axes[2].set_title("Segmentation Watershed")

plt.tight_layout()
plt.show()

# === 8) Sauvegarder le raster segmenté ===
out_path = Path("chemin/vers/mon/CHM_segmente.tif")  # <-- Mets ton chemin cible ici

# Vérifier si le dossier existe
out_path.parent.mkdir(parents=True, exist_ok=True)

with rasterio.open(
    out_path,
    'w',
    driver='GTiff',
    height=labels.shape[0],
    width=labels.shape[1],
    count=1,
    dtype=labels.dtype,
    crs=profile['crs'],
    transform=profile['transform']
) as dst:
    dst.write(labels, 1)

print(f"✅ Raster segmenté sauvegardé : {out_path.resolve()}")
```

```
import geopandas as gpd
from shapely.geometry import Point

# 1. Récupérer les lignes et colonnes (indices image)
rows, cols = coordinates[:, 0], coordinates[:, 1]

# 2. Convertir indices image en coordonnées réelles (x, y)
xs, ys = rasterio.transform.xy(profile['transform'], rows, cols)

# 3. Créer GeoDataFrame avec géométries Points
gdf = gpd.GeoDataFrame(
    geometry=[Point(x, y) for x, y in zip(xs, ys)],
    crs=profile['crs']
)

# 4. Ajouter la hauteur correspondante extraite du CHM
gdf['height'] = chm[rows, cols]

# 5. Sauvegarder en shapefile
out_shp = Path("chemin/vers/mon/sommets_cimes.gpkg")
out_shp.parent.mkdir(parents=True, exist_ok=True)
gdf.to_file(out_shp)

print(f"✅ Sommets vectorisés avec hauteur sauvegardés : {out_shp.resolve()}")
```

DETECTION DES SOMMETS

```
import rasterio
from skimage.feature import peak_local_max
import numpy as np
import geopandas as gpd
from shapely.geometry import Point
import matplotlib.pyplot as plt
from pathlib import Path

# === 1) Charger le CHM ===
chm_path = "/content/chemin/MNH.tif"  # Remplace par ton fichier
with rasterio.open(chm_path) as src:
    chm = src.read(1)
    profile = src.profile

# === 2) Nettoyer les valeurs négatives ===
chm[chm < 0] = 0

# === 3) Détecter les maxima locaux ===
# Ici on utilise le CHM directement !
coordinates = peak_local_max(
    chm,
    min_distance=9,      # distance min entre arbres en pixels (adapter selon résolution)
    threshold_abs=5,     # ignore tout ce qui est < 2m
    labels=(chm > 0)
)

print(f"Points détectés : {len(coordinates)}")

# === 4) Vérifier visuellement ===
plt.figure(figsize=(7, 8))
plt.imshow(chm, cmap='terrain')
plt.scatter(coordinates[:, 1], coordinates[:, 0], c='red', s=10)
plt.title(f"Detected tree tops : {len(coordinates)}")
plt.show()

# === 5) Convertir indices image ➜ coordonnées réelles ===
rows, cols = coordinates[:, 0], coordinates[:, 1]
xs, ys = rasterio.transform.xy(profile['transform'], rows, cols)

# === 6) Créer GeoDataFrame ===
gdf = gpd.GeoDataFrame(
    geometry=[Point(x, y) for x, y in zip(xs, ys)],
    crs=profile['crs']
)

# Ajouter la hauteur
gdf['height'] = chm[rows, cols]

# === 7) Sauvegarder en GeoPackage ===
out_gpkg = Path("/content/sommets_cimes.gpkg")
out_gpkg.parent.mkdir(parents=True, exist_ok=True)

gdf.to_file(out_gpkg, layer="sommets", driver="GPKG")

print(f"✅ Sommets vectorisés sauvegardés : {out_gpkg.resolve()}")

# === 8) Afficher le GeoDataFrame (facultatif) ===
gdf.plot(markersize=1, color="red")
plt.title("Saved canopy points")
plt.show()
```
